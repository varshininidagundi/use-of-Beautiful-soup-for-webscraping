```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=playstation+5&i=fashion&crid=20TF4YO6NG11Z&sprefix=plays%2Cfashion%2C250&ref=nb_sb_ss_ts-doa-p_1_5"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Language':'en-US, en;q=0.5'})
```

```python
webpage = requests.get(URL, headers=HEADERS)
```

```python
#now we need to convert this data into useful
#to do this we use beautifulsoup to convent this data to html
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
#finding link of all elements from amazon
#fecting links as set of tags
links = soup.find_all(id="a-page")
```

```python
links
```

```python
print(links[0].get('href'))
```

```python
tags = soup.find_all("a")
print(tags)
```

```python
tags[76].get("href")
```

```python
only_links = []
base_url = "https://www.amazon.in"

for tag in tags:
    link = tag.get("href")
    if link is not None:
        if not link.startswith("http"):
            link = base_url + link
        only_links.append(link)

print(only_links)
```

```python
df = pd.Series(only_links)
```
